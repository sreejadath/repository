```python
plt.scatter(x_data,y_data)
```

# Histogram

```python
ages = np.random.normal(20,1.5,1000)
ages
```

```python
plt.hist(ages, bins=[ages.min(), 18, 21 , ages.max()])
```

# Pie

```python
lang = ["C++", ".Net" , "Java", "Python","Go"]
vote = [20,30,5,25,45]
explodes = [0,0,0.3,0,0]
plt.pie(vote, labels= lang, explode= explodes
       , autopct = "%.2f%%", pctdistance = 1.4, startangle = 90)
```

# Box Plot

```python
first = np.linspace(0,10,25)
second = np.linspace(10,200,25)
third = np.linspace(200,210,25)
fourth = np.linspace(210,230,25)

data = np.concatenate((first,second,third,fourth))

plt.boxplot(data)
```

# animation

```python
from matplotlib import style
import random
```

```python
head_tails = [0,0]

for _ in range(1000000):
    head_tails[random.randint(0,1)] += 1
    #print(head_tails)
    plt.bar(["head","Tails"],head_tails, color = ["red" , "blue"])
    plt.pause(0.001)
    
plt.show()
```
